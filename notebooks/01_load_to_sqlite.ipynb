# Memuat Data CSV ke SQLite

Notebook ini memuat file CSV mentah dari dataset Olist E-Commerce ke dalam
database SQLite lokal untuk eksplorasi dan analisis berbasis SQL.

**Sumber:** `data/raw/*.csv`  
**Keluaran:** `data/database/olist.db`

## Persiapan

```python
import os
import sqlite3

import pandas as pd
```

```python
DATA_DIR = "../data/raw/"
DB_DIR = "../data/database/"
DB_PATH = os.path.join(DB_DIR, "olist.db")

os.makedirs(DB_DIR, exist_ok=True)
```

## Definisi Tabel

Setiap entri memetakan nama tabel SQLite ke file CSV yang bersesuaian.

```python
TABLES = {
    "customers": "olist_customers_dataset.csv",
    "orders": "olist_orders_dataset.csv",
    "order_item": "olist_order_items_dataset.csv",
    "products": "olist_products_dataset.csv",
}
```

## Muat Data

```python
conn = sqlite3.connect(DB_PATH)

for table_name, file_name in TABLES.items():
    file_path = os.path.join(DATA_DIR, file_name)
    df = pd.read_csv(file_path)
    df.to_sql(table_name, conn, if_exists="replace", index=False)
    print(f"{table_name}: {len(df):,} baris dimuat")

conn.close()
```

## Validasi

Verifikasi bahwa semua tabel berhasil dibuat dan jumlah baris sesuai dengan file CSV sumber.

```python
conn = sqlite3.connect(DB_PATH)

validasi = pd.read_sql_query("""
SELECT 'customers' AS tabel, COUNT(*) AS jumlah_baris FROM customers
UNION ALL
SELECT 'orders', COUNT(*) FROM orders
UNION ALL
SELECT 'order_item', COUNT(*) FROM order_item
UNION ALL
SELECT 'products', COUNT(*) FROM products
""", conn)

conn.close()
validasi
```
